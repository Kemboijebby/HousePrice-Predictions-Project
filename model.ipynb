```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

**EVALUATION AND CALCULATION OF R2 MSE RMSE MAE FROM ACTUAL AND PREDICTED VALUES**

```python
from sklearn import metrics
from sklearn.model_selection import cross_val_score

def cross_val(model):
    pred = cross_val_score(model, X, Y, cv=10)
    return pred.mean()

def print_evaluate(true, predicted):  
    mae = metrics.mean_absolute_error(true, predicted)
    mse = metrics.mean_squared_error(true, predicted)
    rmse = np.sqrt(metrics.mean_squared_error(true, predicted))
    r2_square = metrics.r2_score(true, predicted)
    print('MAE:', mae)
    print('MSE:', mse)
    print('RMSE:', rmse)
    print('R2 Square', r2_square)
    print('__________________________________')
    
def evaluate(true, predicted):
    mae = metrics.mean_absolute_error(true, predicted)
    mse = metrics.mean_squared_error(true, predicted)
    rmse = np.sqrt(metrics.mean_squared_error(true, predicted))
    r2_square = metrics.r2_score(true, predicted)
    return mae, mse, rmse, r2_square
```

**Linear Regression**

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
# Standardize features using StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Create and fit the Linear Regression model
model = LinearRegression()
model.fit(X_train_scaled, Y_train)
```

**Predictions from the model**

```python
Y_pred = model.predict(X_test)
plt.scatter(Y_test,Y_pred)
```

**Residual Histogram**

```python
sns.displot((Y_test -Y_pred), bins=50);
```

**Regression Evaluation Metrics**

```python
Y_Test_pred = model.predict(X_test)
Y_Train_pred = model.predict(X_train)

print('Test set evaluation:\n_____________________________________')
print_evaluate(Y_test, Y_Test_pred)
print('Train set evaluation:\n_____________________________________')
print_evaluate(Y_train,Y_Train_pred)
```

**Results from the model creation**

```python
res_df=pd.DataFrame(data=[["Linear Regression", *evaluate(Y_test,Y_Test_pred),cross_val(LinearRegression())]] ,columns=['Model','MAE','MSE','RMSE','R2 square','Cross validation'])
res_df
```
